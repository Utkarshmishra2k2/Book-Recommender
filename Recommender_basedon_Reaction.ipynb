```python
data_05 = data_04.groupby( "User-ID" ).count()[ "Book-Rating" ]>100
data_05[data_05]
```

```python
data_06 = data_05[data_05].index
```

```python
data_06
```

```python
data_07 = data_04[data_04["User-ID"].isin(data_06)]
```

```python
data_08 = data_07.groupby("Book-Title").count()["Book-Rating"] > 24
data_08[data_08]
```

```python
data_09 = data_08[data_08].index
```

```python
data_10 = data_07[data_07['Book-Title'].isin(data_09)]
```

```python
data_10.drop_duplicates()
```

```python
data_11 = data_10.pivot_table( index = "Book-Title",columns = "User-ID" ,values="Book-Rating" )
```

```python
data_11.fillna(0,inplace = True )
```

```python
data_11
```

```python
data_12 = cosine_similarity(data_11)
```

```python
data_12.shape
```

```python
def function(Book_Name):
    index = np.where(data_11.index==Book_Name)[0][0]
    similar_items = sorted(list(enumerate(data_12[index])),key=lambda x:x[1],reverse=True)[1:4]
    data = []
    for i in similar_items:
        item = []
        temp_df = data_01[data_01['Book-Title'] == data_11.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        data.append(item)
    return data
```

```python
book_name = input(" Please Enter The Book Name : ")
function(book_name)
```
